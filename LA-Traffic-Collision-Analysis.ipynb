Column desctriptions: (As shown on https://data.lacity.org/A-Safe-City/Traffic-Collision-Data-from-2010-to-Present/d5tf-ez2w)

* DR Number: Division of Records Number: Official file number made up of a 2 digit year, area ID, and 5 digits;

* Date Reported: MM/DD/YYYY;

* Date Occurred: MM/DD/YYYY;

* Time Occurred: In 24 hour military time;

* Area ID: The LAPD has 21 Community Police Stations referred to as Geographic Areas within the department. These Geographic Areas are sequentially numbered from 1-21;

* Area Name: The 21 Geographic Areas or Patrol Divisions are also given a name designation that references a landmark or the surrounding community that it is responsible for. For example 77th Street Division is located at the intersection of South Broadway and 77th Street, serving neighborhoods in South Los Angeles.

* Reporting District: A code used in producing reports to group data into geographic sub-areas within an area. 

* Crime Code: Indicates the crime committed. For this dataset - all Crime Code 997;

* Crime Code Description:Defines the Crime Code provided. For this dataset - all Traffic Collision;

* Victim Age: Two character numeric;

* Victim Sex: F - Female M - Male X - Unknown;

* Victim Descent: Descent Code: A - Other Asian B - Black C - Chinese D - Cambodian F - Filipino G - Guamanian H - Hispanic/Latin/Mexican I - American Indian/Alaskan Native J - Japanese K - Korean L - Laotian O - Other P - Pacific Islander S - Samoan U - Hawaiian V - Vietnamese W - White X - Unknown Z - Asian Indian;

* Premise Code: The type of structure or location where the incident took place;

* Premise Description: Defines the Premise Code provided;

* Address: Street address of crime incident rounded to the nearest hundred block to maintain anonymity;

* Cross Street: Cross Street of rounded Address;

* Location: The location where the crime incident occurred. Actual address is omitted for confidentiality. XY coordinates reflect the nearest 100 block;

Creating the first table for location of the accident. This table includes the area name (of the accident), type/description of the accident, remise description, address (including street name of the accident), cross street, and geo location. 

```python
%%sql
DROP TABLE IF EXISTS location;

CREATE TABLE location (
    location_key SERIAL PRIMARY KEY,
    area_name VARCHAR(64),
    crime_code_descript VARCHAR(64),
    premise_descript VARCHAR(64),
    address VARCHAR(64),
    cross_street VARCHAR(64),
    geo VARCHAR(64)
);
```

Setting all the null values in the inputs into location into the subscript 'NULL' so they are still evaluated and inputted into the table. Otherwise we were having issues with creating a key value for all those rows with null values in thse columns. 

```python
%%sql
UPDATE main 
SET area_name = 'NULL' WHERE area_name IS NULL;
UPDATE main 
SET premise_descript = 'NULL' WHERE premise_descript IS NULL;
UPDATE main 
SET address = 'NULL' WHERE address IS NULL;
UPDATE main 
SET cross_street = 'NULL' WHERE cross_street IS NULL;
UPDATE main 
SET location_geo = 'NULL' WHERE location_geo IS NULL;
UPDATE main 
SET crime_code_descrpit = 'NULL' WHERE crime_code_descrpit IS NULL;
```

The following cells add the original unique values into the new table, and add in the key for the main table. 

```python
%%sql
INSERT INTO location (area_name, crime_code_descript, premise_descript, address, cross_street, geo)
SELECT DISTINCT area_name, crime_code_descrpit, premise_descript, address, cross_street, location_geo FROM main
```

```python
%%sql
ALTER TABLE main
ADD COLUMN location_key INTEGER,
ADD CONSTRAINT fk_location_key         
    FOREIGN KEY (location_key)
    REFERENCES location (location_key);
```

```python
%%sql
DROP INDEX IF EXISTS idx_location;

CREATE INDEX idx_location ON location (area_name, crime_code_descript, premise_descript, address, cross_street, geo);
```

```python
%%sql
UPDATE main
SET location_key = location.location_key
FROM location
WHERE main.area_name = location.area_name 
    AND main.crime_code_descrpit = location.crime_code_descript
    AND main.premise_descript = location.premise_descript
    AND main.address = location.address
    AND main.cross_street = location.cross_street
    AND main.location_geo = location.geo;
```

Creating the new table for time. 

```python
%%sql
DROP TABLE IF EXISTS time;

CREATE TABLE time (
    time_key SERIAL PRIMARY KEY,
    day CHAR(10),
    year INTEGER,
    month_of_year_str VARCHAR(12),
    month_of_year INTEGER,
    day_of_month INTEGER,
    day_of_week_str CHAR(9),
    day_of_week INTEGER,
    is_weekend BOOLEAN,
    is_weekday BOOLEAN, 
    hour INTEGER
);
```

Note: The following code (for inserting the data into the time table) is similar to that shown in class, and used in class exercises. We made a few modifications to adjust to our data. 

```python
%%sql
INSERT INTO time (day, year, month_of_year_str, month_of_year, day_of_month, 
                  day_of_week_str, day_of_week, is_weekend, is_weekday, hour)
SELECT DISTINCT day AS day, 
    CAST(TO_CHAR(date_occurred, 'YYYY') AS INTEGER) AS year,
    TO_CHAR(date_occurred, 'Month') AS month_of_year_str,
    CAST(TO_CHAR(date_occurred, 'MM') AS INTEGER) AS month_of_year,
    CAST(TO_CHAR(date_occurred, 'DD') AS INTEGER) AS day_of_month,
    TO_CHAR(date_occurred, 'Day') AS day_of_week_str,
    CAST(TO_CHAR(date_occurred, 'D') AS INTEGER) AS day_of_week,
    CASE WHEN CAST(TO_CHAR(date_occurred, 'D') AS INTEGER) IN (1, 7) 
        THEN TRUE
        ELSE FALSE
    END AS is_weekend,
    CASE WHEN CAST(TO_CHAR(date_occurred, 'D') AS INTEGER) NOT IN (1, 7) 
        THEN TRUE
        ELSE FALSE
    END AS is_weekday,
    hour AS hour
FROM main;
```

Which areas have a high risk of collisions? (Top 3)

```python
%%sql
SELECT location.area_name AS Area, count(main.dr_number) AS Total_Collisions
FROM main
JOIN location ON location.location_key = main.location_key
WHERE location.area_name IS NOT NULL
GROUP BY location.area_name
ORDER BY Total_Collisions DESC
Limit 3;
```

The 77th Street Area has the most collisions. 

Which streets are of the most risk in the highest risky areas of LA?

```python
%%sql
SELECT location.area_name AS Area, location.address as Street_Name, count(main.dr_number) AS Total_Collisions
FROM main
JOIN location ON location.location_key = main.location_key
WHERE location.area_name IS NOT NULL AND location.address IS NOT NULL
GROUP BY location.area_name, location.address
ORDER BY Total_Collisions DESC
Limit 10;
```

It appears that Hollywood regions' Sunset Boulevard has had the most traffic collisions in LA. This is not necessasily that surprising considering how many actors/industry people live/work there, and are likely involved in many accidents. 

What time has the highest requencies of traffic collsions?

```python
%%sql
SELECT time.hour AS Hour, count(main.dr_number) AS Total_Collisions
FROM main
JOIN time ON time.time_key = main.time_key
WHERE time.hour IS NOT NULL
GROUP BY time.hour
ORDER BY Total_Collisions DESC
Limit 5;
```

It appears that teh 5 o'clock hour in the evenings is the most common time for accidents to happen. It is followed by 3 o'clock, and 6 o'clock. It appears that there are more traffic collisions happening in the evenings as the top 5 hours shown in the table above are all after 12pm. 

Is male or female more likely to get involved into the traffic collsions?

```python
%%sql
SELECT victim.sex AS Sex, count(main.dr_number) AS Total_Collisions
FROM main
JOIN victim ON victim.victim_key = main.victim_key
GROUP BY victim.sex
ORDER BY Total_Collisions DESC
Limit 3;
```

Majority of victims are men. 'X' in this case means that the victims' sex was not known. 

What are the top premise descriptions has the highest risk in the most dangerous area?

```python
%%sql
SELECT location.premise_descript AS Area, count(main.dr_number) AS Total_Collisions
FROM main
JOIN location ON location.location_key = main.location_key
WHERE location.premise_descript IS NOT NULL
GROUP BY location.premise_descript
ORDER BY Total_Collisions DESC
Limit 5;
```

As would be expected, majority of traffic collisions, by an abnormally high margin, take place on 'Streets', followed by Parking Lots, and sidewalks (Which was a little bit of a surprise to us - considering how many there are. 

What are the top victim descents?

```python
%%sql
SELECT victim.descent AS Descent, count(main.dr_number) AS Total_Collisions
FROM main
JOIN victim ON victim.victim_key = main.victim_key
GROUP BY victim.descent
ORDER BY Total_Collisions DESC
Limit 5;
```

What are the ages for the highest number of accidents?

```python
%%sql
SELECT victim.age AS Age, count(main.dr_number) AS Total_Collisions
FROM main
JOIN victim ON victim.victim_key = main.victim_key
GROUP BY victim.age
ORDER BY Total_Collisions DESC
LIMIT 5;
```

Which years have had more accidents? 

```python
%%sql
SELECT time.year AS Year, count(main.dr_number) AS Total_Collisions
FROM main
JOIN time ON time.time_key = main.time_key
GROUP BY time.year
ORDER BY Total_Collisions DESC;
```

It appears that collisions have been increasing since 2010. 2013 is the only exception. Although it shows 2018 has less collisions than 2015 (currently), this data is only until end of November, and the expectation is that it would increase to above 2017, by the end of the year, especially as more and more people travel for the holidays. This can also be evaluated by the next question examining the distribution of collisions by month. 

Which months have had more accidents? 

```python
%%sql
SELECT time.month_of_year_Str AS Month, count(main.dr_number) AS Total_Collisions
FROM main
JOIN time ON time.time_key = main.time_key
GROUP BY Month
ORDER BY Total_Collisions DESC;
```

It appears that October has the highest number of traffic collisions. On the other hand, December has the least number of collisions. Unlike our initial thought, December, as shown by the table, likely would not lead 2018 to have more collisions than 2017. This shows that LA Police recent safety processes are working effectively. 

Which days of the week have had more accidents? 

```python
%%sql
SELECT time.day_of_week_str AS DayOfWeek, count(main.dr_number) AS Total_Collisions
FROM main
JOIN time ON time.time_key = main.time_key
GROUP BY time.day_of_week_str
ORDER BY Total_Collisions DESC;
```

Fridays are most common day of the week causing accidents, followed by Saturday and Thursday. This makes sense as there are more people on the road on weekends. However, the accidents are pretty evenly divided amongst all the days of the weeks. 

### More Complex Analysis/Questions

At what hour of the day of the week results in most accidents? 

```python
%%sql
SELECT time.day_of_week_str AS DayOfWeek, time.hour AS Hour, count(main.dr_number) AS Total_Collisions
FROM main
JOIN time ON time.time_key = main.time_key
GROUP BY time.day_of_week_str, time.hour
ORDER BY Total_Collisions DESC
Limit 5;
```

Fridays at 5 pm are the most common time for accidents, followed by the same day of the week at 3 pm. Also, as shown through the sql table earlier, weekends and evening times are most common for traffic collisions. 

What victim gender is more common in what districts at what hours of the days of the week? (Combining all three key tables)

```python
%%sql
SELECT victim.sex, location.area_name AS AreaName, time.day_of_week_str AS DayOfWeek, time.hour AS Hour, count(main.dr_number) AS Total_Collisions
FROM main
JOIN time ON time.time_key = main.time_key
JOIN location ON location.location_key = main.location_key
JOIN victim ON victim.victim_key = main.victim_key
GROUP BY time.day_of_week_str, time.hour, location.area_name, victim.sex
ORDER BY Total_Collisions DESC
Limit 5;
```

It appears that men in the Hollywood region are more likely to be involved in an accident at 2 am. That region, at 2 am, is also very dangerous on Sunday nights.

### Analysis with weather data

Starting with a few basic analysis questions:

At what temperature are the accidents most common?

```python
%%sql
SELECT weather.temp AS Temp, count(main.dr_number) AS Total_Collisions
FROM main
JOIN weather ON weather.weather_key = main.weather_key
GROUP BY weather.temp
ORDER BY Total_Collisions DESC
Limit 5;
```

So, it appears that 63F has the highest accidents, however, the temperature in LA, CA, doesn't really change much as observed by how close all the total_collisions values are shown above. Maybe looking at wind_speed or precip might work better. 

```python
%%sql
SELECT weather.wind_speed AS windspeed, count(main.dr_number) AS Total_Collisions
FROM main
JOIN weather ON weather.weather_key = main.weather_key
GROUP BY weather.wind_speed
ORDER BY Total_Collisions DESC
Limit 5;
```

It appears that wind is not really prevelent in this reagion as the most accidents have occurred when the wind speed is 0. Let's check precipitation. 

```python
%%sql
SELECT weather.precip AS Precipitation, count(main.dr_number) AS Total_Collisions
FROM main
JOIN weather ON weather.weather_key = main.weather_key
GROUP BY weather.precip
ORDER BY Total_Collisions DESC
Limit 5;
```

As expected, LA, CA, doesn't really have much rain, and essentially all the values shown here (Except None) are 0 in value, showing that rain/precipitation doesn't cause any accidents in this region either. 

So, weather, in measurable metrics, is not really a factor in characterizing/predicting accidents.

Finally, lets look at the characterized 'weather' descriptions available and how they different in accidents happening. These descriptions are provided by the NOAA dataset. 

```python
%%sql
SELECT weather.weather AS Temp, count(main.dr_number) AS Total_Collisions
FROM main
JOIN weather ON weather.weather_key = main.weather_key
GROUP BY weather.weather
ORDER BY Total_Collisions DESC
Limit 5;
```

So, this shows that Scattered (SCT) Weather Type and 0.10% of rain (FEW) are the two most likliest weather conditions for accidents to take place in.  

Doing similar analysis like last time, trying to connect all 5 tables within the database together. 

What victim gender is more common in what districts, under what weather conditions, at what hours of the days of the week? (Combining all key tables)

```python
%%sql
SELECT victim.sex, location.area_name AS AreaName, weather.weather, time.day_of_week_str AS DayOfWeek, time.hour AS Hour, count(main.dr_number) AS Total_Collisions
FROM main
JOIN time ON time.time_key = main.time_key
JOIN location ON location.location_key = main.location_key
JOIN victim ON victim.victim_key = main.victim_key
JOIN weather ON weather.weather_key = main.weather_key
GROUP BY time.day_of_week_str, time.hour, location.area_name, weather.weather, victim.sex
ORDER BY Total_Collisions DESC
Limit 5;
```

It appears that men in the 77th Street, during few (10% chance of rain) weather condition, at 4 pm on Tuesdays have the most collisions. However, there doesn't appear to be much difference between the results showing that all these factors may not be influencing the collisions together as much as we had taught earlier. 

These results are also different than those seen before, where Hollywood region was more likely to be involved in an accident at 2 am on Saturdays and Sundays.
